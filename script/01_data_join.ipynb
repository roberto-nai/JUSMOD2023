```python
import pandas as pd
import os
import csv
```

```python
# GLOBALS
data_dir = "data"
country_code = "ES" # <-- INPUT: desired country [IT, FR, ES]
cae_type = "3" # <-- INPUT: CAE_TYPE "3 Regional or local authority"
file_ted = "_Export_OpenDataCFC_year2016_2022_"+country_code+".csv" # country_code tender notice
file_awd = "_Export_OpenDataCAN_year2016_2022_"+country_code+".csv" # country_code award notice
```

```python
# Reading the CSV ted
path_data = os.path.join(data_dir, file_ted)
dt = {'ID_NOTICE_CN':object, 'FUTURE_CAN_ID':object, 'FUTURE_CAN_ID_ESTIMATED':object, 'MAIN_CPV_CODE_GPA':object}
df_ted = pd.read_csv(path_data, sep = ";", dtype=dt, low_memory=False)
```

```python
# df_ted['FUTURE_CAN_ID'].dtypes
```

```python
# Get tenders of CAE_TYPE "3 Regional or local authority"
df_ted_local = df_ted[df_ted['CAE_TYPE'] == cae_type]
```

```python
file_out = "_Export_OpenDataCFC_year2016_2022_"+country_code+"_Type_"+cae_type+".csv"
path_data = os.path.join(data_dir, file_out)
df_ted_local.to_csv(path_data, sep = ";", index = False)
```

```python
# Reading the CSV award
path_data = os.path.join(data_dir, file_awd)
dt = {'ID_NOTICE_CAN':object, 'ID_AWARD':object, 'ID_LOT_AWARDED':object, 'MAIN_CPV_CODE_GPA':object}
df_awd = pd.read_csv(path_data, sep = ";", dtype=dt, low_memory=False)
```

```python
# Get tenders of CAE_TYPE "3 Regional or local authority"
df_awd_local = df_awd[df_awd['CAE_TYPE'] == cae_type]
```

```python
file_out = "_Export_OpenDataCAN_year2016_2022_"+country_code+"_Type_"+cae_type+".csv"
path_data = os.path.join(data_dir, file_out)
df_awd_local.to_csv(path_data, sep = ";", index = False)
```

```python
# Join  
```

```python
# Before the join, exclude duplicated columns
cols_to_use = df_awd_local.columns.difference(df_ted_local.columns) # get the columns tha are only in df_awd_local
```

```python
df_join = pd.merge(left = df_ted_local, right = df_awd_local[cols_to_use], 
                   left_on='FUTURE_CAN_ID', right_on='ID_NOTICE_CAN', 
                   how = 'inner')
```
